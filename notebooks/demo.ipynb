# EMPA tools package

In the following we will demonstrate the empatools package. It is currently capable of doing just few impressive things:

```python
from empatools import *
```

```python
dt = Oxides.example_data()
dt
```

```python
g = dt.get_sample('-g-')
g
```

```python
grt = mindb.Garnet_Fe2()
apfu = g.apfu(grt)
apfu
```

```python
apfu.endmembers()
```

```python
f = dt.get_sample('-pl-')
f
```

```python
plg = mindb.Feldspar()
apfu = f.apfu(plg)
apfu
```

```python
apfu.endmembers()
```

```python
gp = Oxides.example_data(sample='grt_profile')
em = gp.apfu(grt).endmembers()
minplot.plot_grt_profile(em, percents=True)
```
